```python
#Importing the necessary libraries
import pandas as pd
import pydotplus
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_graphviz
from six import StringIO
from IPython.display import Image
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
```

```python
#I am importing the dataset from my computer with the address below
heart = pd.read_excel(r'C:\Users\dobimuyiwa\Documents\Exercise Files\Refresher\heart.xlsx')
print(heart.head())
```

```python
heart_column = ['sex', 'cp', 'fbs', 'restecg', 'exang']

# I need to differentiate my features from my labels
features = heart.drop('target', axis=1)  #This means everyother column except the 'target' column is a feature
labels = heart['target']   #This means 'target' is a label

print(heart.tail())
```

```python
#I need to split my dataset into training and test dataset
# x = features, y = labels. I am setting aside 80% for training set and 20% for test set.
x_train, x_test, y_train, y_test = train_test_split(features, labels, test_size=0.2)
```

```python
#build the pipeline
pipe = Pipeline([
    ('scale', StandardScaler()), #The idea of standard scaler is to transform your data such that its
                                #distribution wil have a mean = 0 and SD = 1
    ('Dct', DecisionTreeClassifier(criterion='gini')),
])

pipe.fit(x_train, y_train)
```

```python
#Evaluate the model's prediction
y_pred = pipe.predict(x_test)
print(y_pred)
print(type(y_pred))
y_expect = y_test.to_numpy()
print(y_expect)
print(type(y_expect))
print(len(y_expect))
```

```python
print(metrics.classification_report(y_expect, y_pred))
```

```python
print(confusion_matrix(y_test, y_pred))
```
