```python
#DAY 2 PROJECT -> WITH TAXES
print("Welcome to the tip calculator!", '\n')
bill = input("What was the total bill? ")
#I have decided to add GST into the calculation
bill_gst = int(bill) * (13/100) 
Tbill = float(bill_gst) + int(bill)
print(f"Your total bill is: ${Tbill} plus tax", '\n')
#Calculating the tip
tip = input("What percentage tip would you like to give? 10%, 15%, or 20%? ")
tip_percent = int(tip) / 100
Tbill_times_tip = float(Tbill) * float(tip_percent)
Tbill_plus_tip = float(Tbill) + Tbill_times_tip
#Calculate bill per person
persons = input("How many people would split the bill? ")
bill_person = Tbill_plus_tip / int(persons)

print("........Total Amount Per Person would be.......", '\n')
print(f"Each person would pay: ${bill_person}", '\n')
print("........Thank you so much for your kindness.......")

```

```python
#DAY 2 PROJECT -> WITHOUT TAXES
print("Welcome to the tip calculator!", '\n')
bill = input("What was the total bill? ")
print(f"Your total bill is: ${bill}", '\n')
#Calculating the tip
tip = input("What percentage tip would you like to give? 10%, 15%, or 20%? ")
tip_percent = int(tip) / 100
bill_times_tip = float(bill) * float(tip_percent)
bill_plus_tip = float(bill) + bill_times_tip
#Calculate bill per person
persons = input("How many people would split the bill? ")
bill_person = bill_plus_tip / int(persons)

print("........Total Amount Per Person...................", '\n')
Bill_person = "{:.2f}".format(bill_person) #returns answer in 2 decimal places
print(f"Each person would pay: ${Bill_person}", '\n')
print("........Thank you so much for your kindness.......")
```
