Understanding date-time object
1. use the built-in Python datetime module -> from datetime import datetime, date, time, timedelta

```python
from datetime import datetime, date, time, timedelta

dtm = datetime(2013, 5, 17, 16, 21, 18) #using datetime type construction method
print("dtm is", dtm)
print(dtm.year, dtm.month, dtm.day, dtm.hour, dtm.minute, dtm.second)

dt = datetime.strptime("20130517@162118", "%Y%m%d@%H%M%S") #convert string to datetime: strptime -> string-parsed-to-datetime
print("dt is", dt)


dtt = dtm.strftime("%m/%d/%Y %H:%M") #read more bout this.... Convert datetime to string
print("dtt is", dtt)
```

```python
#timedelta
d1 = date(2022, 1, 27)
d2 = date(2022, 1, 31)
print(d2 - d1) #returns a timedelta object

delta = timedelta(hours=72, minutes=0, seconds=0)
olddt = datetime(2022, 1, 27, 8, 21)
newdt = olddt + delta
print(newdt)
```

Set Objects and operations
1. Set is an unordered collection of unique elements. It can be used on a list and a tuple.
2. Use set() function to generate a set
3. Use myset.add('newitem') to add new item to your set
4. Use myset.discard('olditem') to remove old item from your set

x in s -> test x for membership in s
x not in s -> test x for non-membership in s
s.issubset(t) or s<=t -> test whether every element in s is in t
s.issuperset(t) or s>=t -> test whether every element in t is in s
s.union(t) or s | t -> new sets with elements from both s and t or s.update(t)
s.intersection(t) or s & t -> new set with elements common to s and t or s.intersection_update(t)
s.difference(t) or s - t -> new set with elements in s but not in t
s.symmetric_difference(t) or s ^ t -> new set with elements in either s or t but not both.
s.copy() -> generates a copy of s

Variable Scope
1. Global scope is used to define the global scope inside a function

```python
name = 'Mary' #Initial global
age = 25 #Global

def identity():
    global name #This automatically changes the variable called name to whatever was declared afterwards. In this case, it's John.
    age = 18 #Local
    name = 'John' #New Global
identity()
print('My name is ' + name + ' and I am ' + str(age))

def print_me(name, age):
    print('My name is ' + name + ' and I am ' + str(age))
    return(name+','+str(age))
x = print_me('Larry', 19)
print(x)

def outfunc(a, b):
    x=a+b #a=5, b=2, x=7
    c=1
    def infunc(p, q): #p=a, q=b
        return (p - q + c) #5-2+1 = 4
    y = infunc(a, b) #a=5, b=2. So therefore, y = 4
    return(x*y) #Final answer would be 7 * 4 = 28
print('The answer is ' + str(outfunc(5, 2)))

def score(myanswer, answer=('a', 'c', 'd', 'a', 'b')):
    score=0
    j=0
    while (j<len(answer)):
        if(answer[j-1] == myanswer[j-1]):
            score+=1
        j+=1
    print('Your score is ' +str(score) +' for the test')
    #return(score)
score(('a', 'a', 'd', 'c', 'b'))

#Lamba function -- single line function
areacircle = lambda r: 3.1416 * r * r
print('The area of a circle with r-5.2 is: ', areacircle(5.2))
```
