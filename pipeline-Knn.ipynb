```python
import pandas as pd
import numpy as np
import sklearn
import scipy
import urllib
import matplotlib.pyplot as plt

from pylab import rcParams
from sklearn import neighbors
from sklearn import preprocessing
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
```

```python
plt.style.use('seaborn-whitegrid')
```

```python
cars = pd.read_csv(r'C:\Users\dobimuyiwa\Documents\Exercise Files\Ex_Files_Python_Data_Science_EssT_Pt2\Exercise Files\Data\mtcars.csv')
cars.columns = ['car_names', 'mpg', 'cyl', 'disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']

cars.head()
```

```python
x_prime = cars[['mpg', 'disp', 'hp', 'wt']].values
y = cars.iloc[:,9].values #This returns the values for the 'am' column
print('y is', y)
```

```python
#Scale your variables/features before creating the model
x = preprocessing.scale(x_prime)
print(x)

"""""
The preprocessing.scale() algorithm puts your data on one scale. This is helpful with largely sparse datasets. 
In simple words, your data is vastly spread out. For example the values of X maybe like so:

X = [1, 4, 400, 10000, 100000]

The issue with sparsity is that it very biased or in statistical terms skewed. 
So, therefore, scaling the data brings all your values onto one scale eliminating the sparsity. 
"""""
```

```python
#split and train data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=17)
```

```python
#build the pipeline
pipe = Pipeline([
    ('scale', StandardScaler()), #The idea of standard scaler is to transform your data such that its
                                #distribution wil have a mean = 0 and SD = 1
    ('knn', KNeighborsClassifier()),
])

pipe.fit(x_train, y_train)
```

```python
#Evaluate the model's prediction
y_pred = pipe.predict(x_test)
print(y_pred)
y_expect = y_test
print(y_expect)
```

```python
print(metrics.classification_report(y_expect, y_pred))
```
